```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

browser = webdriver.Chrome('./bin/chromedriver')

#url login page
browser.get('https://myjobstreet.jobstreet.com.my/home/login.php?site=my&language_code=3&go=JOB-ADS&nrfr=1')

#inspect element on login page to to find id/username
user = browser.find_element_by_id('login_id')
user.send_keys('###########@yahoo.com')

#same for pw
pw = browser.find_element_by_id('password')
pw.send_keys('#########')

#send user name &keys
pw.send_keys(Keys.RETURN)

#get intended webpage
browser.get('https://www.jobstreet.com.my/en/job-search/campus/job-vacancy.php?area=1&option=1&location=50700&job-type=1%7C2%7C4%7C8%7C32&job-source=1%2C64&classified=1&job-posted=0&src=45&pg=1&sort=1&order=0&srcr=12&vtype=list')

#get page source
code = browser.page_source

```

```python
#now can proceed using soup as usual
from bs4 import BeautifulSoup
soup = BeautifulSoup(code,'html.parser')
```

```python
#inspect element find class which contain data 
soup_list = soup.findAll(class_ = 'rRow')
soup_list[0] #test first entry
```

```python
#keep testing first entry

# print(soup_list[1].find(class_='colTitle').get_text())
# print(soup_list[0].find(class_='colComp').get_text())
# print(soup_list[0].find(class_='colSal').get_text())
# print(soup_list[0].find(class_='smalldesc').get_text())


#company list to extract industry info
Company1 = [soup_list.find(class_='colComp') for soup_list in soup_list]

```

```python
#choosen list of atributes

Job_Title = [soup_list.find(class_='rRowTitle').get_text() for soup_list in soup_list]
Company = [soup_list.find('h3').get_text() for soup_list in soup_list]
Job_Category = [soup_list.find(class_='smalldesc').get_text() for soup_list in soup_list]
Industry = [Company1[items].find("div", attrs={"class":"smalldesc"}).text.replace('Industry:','') for items in range(len(Company1))]
Salary = [soup_list.find(class_='colSal').get_text().replace('\n','').replace('MYR','').replace(',','') for soup_list in soup_list]
```

```python
#create pandas df

import pandas as pd

Jobstreet = pd.DataFrame({
    'Job_Title':Job_Title,
    'Company':Company,
    'Job_Category':Job_Category,
    'Industry': Industry,
    'Salary':Salary,  
    
})
```
